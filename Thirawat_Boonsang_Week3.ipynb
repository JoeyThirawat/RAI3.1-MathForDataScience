```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
# exercise a

A = [3,4]
B = [5,2]

x_pos = A[0]
y_pos = A[1]
x_dir = abs(x_pos - B[0])
y_dir = B[1] - y_pos

# length of line segment
a1 = abs(x_pos - B[0])
a2 = abs(y_pos - B[1])
len_line_seg = math.sqrt(a1**2+a2**2)
print(f"Length of line segment = {len_line_seg}")

# plot a vector on the graph
ax = plt.gca()
ax.quiver(x_pos, y_pos, x_dir, y_dir, angles='xy', scale_units='xy', scale=1, color=['red'])

ax.axis([-1,10,-1,10])

plt.show()
```
