```python
import sympy as sp
import math
from sympy.plotting import plot
from treelib import Node, Tree
```

image.png

```python
#1
#A
tree = Tree()
tree.create_node("Components", "All") #root
#C
tree.create_node("Machine A 40%", "A", parent="All")
tree.create_node("Fail 3%", "AF", parent="A")
tree.create_node("Success 97%", "AS", parent="A")
#B
tree.create_node("Machine B 35%", "B", parent="All")
tree.create_node("Fail 2%", "BF", parent="B")
tree.create_node("Success 98%", "BS", parent="B")
#C
tree.create_node("Machine C 25%", "C", parent="All")
tree.create_node("Fail 1%", "CF", parent="C")
tree.create_node("Success 99%", "CS", parent="C")

tree.show()
```

```python
#1B
faulty = (.4*.03)+(.35*.02)+(.25*.01)
print("Probability that components will be faulty is", faulty)

#1C
fault_machine_C = (.25*.01)/faulty
print("Probability it was produced by machine C is",fault_machine_C)
```

```python
# 2A
tree = Tree()
tree.create_node("population", "pop") #root
# Disease
tree.create_node("Disease 2%", "d", parent="pop")
tree.create_node("Detected + 95%", "dt", parent="d")
tree.create_node("Not Detected - 5%", "ndt", parent="d")

# No disease
tree.create_node("No Disease 98%", "nd", parent="pop")
tree.create_node("Detected + 3%", "nd_dt", parent="nd")
tree.create_node("Not Detected - 97%", "nd_ndt", parent="nd")

tree.show()
```

```python
#2B
pos = (.02*.95)+(.98*.03)
print("Probability of selected at random from pop and get positive is",pos)

#2C
prob = (.98*.03)/pos
print("Person with positive test does not have the disease is", prob)
```

```python
#3
a, b, x, var= sp.symbols('a, b, x, variance')
ex = 1/(b-a)
print("Uniform distribution function")
display(ex)
print("Where a <= x <= b, otherwise = 0")
print("====================")

print("\nExpectation (mean)")
exMean = (b + a)/2
display(exMean)
mean = exMean.subs({a:0, b:25})
print(f"Mean: {float(mean):.2f}")
print("====================")

print("\nVariance")
exVar = ((b-a)**2)/12
display(exVar)
variance = exVar.subs({a:0, b:25})
print(f"Variance: {float(variance):.2f}")
print("====================")

print("\nStandard deviation")
exStd = sp.sqrt(var)
display(exStd)
std = sp.sqrt(variance)
print(f"Standard deviation: {float(std):.2f}")
print("====================")
```
