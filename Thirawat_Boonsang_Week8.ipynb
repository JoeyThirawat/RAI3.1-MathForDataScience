```python
import sympy as sp
from sympy import *
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# A
x,y = sp.symbols('x y')

line_y = 4
curve = x*(5-x)
xpoint = solve(Eq(curve, line_y), x)
ypoint = np.ones(len(xpoint))*line_y

print(f"Intersection Points is ({xpoint[0]}, {int(ypoint[0])}), ({xpoint[1]}, {int(ypoint[1])})")
plot(curve, line_y, xlim=(0,5), ylim=(0,8), markers=[{'args':[xpoint[0], curve.subs(x,xpoint[0]), 'ro']},
                                                    {'args':[xpoint[1], curve.subs(x,xpoint[1]), 'ro']}])
```

```python
# B
A = integrate(curve,(x,xpoint[0],xpoint[1]))
print(f"Shaded area is {A}")

x_axis = np.arange(0, 5, 0.01)
y_axis = x_axis*(5-x_axis)

plt.plot(x_axis, y_axis)
plt.axhline(4)
plt.fill_between(x_axis, y_axis, 4, where=(x_axis>=1) & (x_axis<=4) , color='orange')
```

```python
# B
x = sp.symbols('x')
midpoint = np.array([2.75,6.5,9,13,20.5,38])
frequency = np.array([20.0,16.0,18.0,25.0,15.0,10.0])
n = int(frequency.sum())

# Q2
q2pos = n/2 
print(f"Median position : {q2pos} th")
print("=> between 7.5 - 10.5 hours")
eq1 = (x-7.5) / (q2pos - (20+16))
eq2  = (10.5-7.5)/((20+16+18)-(20+16))
q2val = solve(Eq(eq1,eq2),x)

print("")
print(f"Q2 position or Median: {q2pos}")
print(f"Median(Q2) val: {round(q2val[0],2)}")
print("")

# Q1
q1pos = n/4 
print(f"Q1 position : {q1pos} th")
print("it is between 5.5 - 7.5")
eq1 = (x-5.5)/(q1pos - 20)
eq2  = (7.5-5.5)/((20+16)-20)
q1val = solve(Eq(eq1,eq2),x)
print(f"Q1 val : {round(q1val[0],2)}")
print("")

# Q3
q3pos = 3/4*n
print(f"Q3 position : {q3pos}")
print("it is between 10.5 - 15.5")
eq1 = (x-10.5)/(q3pos - (20+16+18))
eq2  = (15.5-10.5)/((20+16+18+25)-(20+16+18))
q3val = solve(Eq(eq1,eq2),x)
print(f"Q3 val : {round(q3val[0],2)}")

## Range
print("\n")
print(f"Range = Q3-Q1 = {round(q3val[0] - q1val[0],3)}")
```

```python
# C
mf = midpoint * frequency
total = np.sum(mf)
mean = total / n
print(f"Mean value = {round(mean,2)} hrs")

x_min_mean = midpoint - mean
square = x_min_mean**2
times = frequency*square

var = times / (n-1)
var_res = var.sum()
standard_dev = sqrt(var_res)
print(f"Standard deviation = {round(standard_dev,2)}")
```
