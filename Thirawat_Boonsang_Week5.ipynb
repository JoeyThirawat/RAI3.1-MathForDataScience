### 1. Suppose we have data points given by (Xn, Yn) = (0, 1), (1, 3), (2, 3), and (3, 4). By solving the normal equations, fit the data by the line Y = a + bX

```python
X = np.array([0,1,2,3])
y = np.array([1,2,3,4])

plt.plot(X, y,'o')
plt.xlabel('X')
plt.ylabel('y')
```

```python
X_b = np.c_[np.ones((4, 1)), X]
print(X_b)
```

```python
b = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

y_p = b[0] + b[1]*X # plot the y value
# Y = a + bX

plt.plot(X, y_p, '-r')
plt.plot(X,y,'o')
plt.xlabel('X')
plt.ylabel('y')
plt.grid()
plt.show()
```

### 2. To prepare the Bangkok Governer election, a political group hired a public relations firm to promote its candidate in three way; Telephone, Social Media and Roadshow. The cost per ads is given as:
### Telephone 0.04 ฿; Social Media 0.05 ฿; Roadshow 0.1 ฿ per contact. The number of contacts of each type made in first 2 potential postcode A and B is given by

### Find the total amount spent by the group in the two postcode A and B

```python
A = np.array([1000, 500, 5000])
B = np.array([3000, 1000, 10000])
cost = np.array([4, 5, 10])

A_amt = cost*A # Amount of A
B_amt = cost*B # Amount of B

print("Postcode A =", A_amt)
print("Postcode B =", B_amt)
```

```python
A_amt_final = np.sum(A_amt)
print("The total amount spent by the group in the postcode A is", A_amt_final, "Baht")
```

```python
B_amt_final = np.sum(B_amt)
print("The total amount spent by the group in the postcode B is", B_amt_final, "Baht")
```
