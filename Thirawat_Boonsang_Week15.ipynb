```python
# sales in each continent

NA = df['North American Sales'].sum()
EU = df['EU Sales'].sum()
JP = df['Japan Sales'].sum()
oth = df['Other Sales'].sum()
sum = df['World Sales'].sum()

labels = 'North America', 'Europe', 'Japan', 'other'
sizes = [NA, EU, JP, oth]
plt.pie(sizes, labels=labels, autopct='%.2f')
plt.axis('equal')
plt.show()
```

```python
# Japan aspect
genre_JP = df[["Genre", "Japan Sales"]]
genre_JP_sum = genre_JP.groupby(['Genre']).sum()
genre_JP_sum.plot(kind = 'pie', y ='Japan Sales', autopct='%.1f', figsize =(10, 10))
```

```python
# Europe aspect
genre_EU = df[["Genre", "EU Sales"]]
genre_EU_sum = genre_EU.groupby(['Genre']).sum()
genre_EU_sum.plot(kind = 'pie', y ='EU Sales', autopct='%.1f', figsize =(10, 10))
```

```python
# north American aspect
genre_NA = df[["Genre", "North American Sales"]]
genre_NA_sum = genre_NA.groupby(['Genre']).sum()
genre_NA_sum.plot(kind = 'pie', y ='North American Sales', autopct='%.1f', figsize =(10, 10))
```

```python
# other aspect
genre_other = df[["Genre", "Other Sales"]]
genre_other_sum = genre_other.groupby(['Genre']).sum()
genre_other_sum.plot(kind = 'pie', y ='Other Sales', autopct='%.1f', figsize =(10, 10))
```

```python
# genre of game list

df['Genre'].value_counts()
```

```python
plt.figure(figsize=(15, 10))
sns.countplot(x="Genre", data=df, order = df['Genre'].value_counts().index)
plt.xticks(rotation=90);
```

```python
# the increasing of year that games has relased

plt.figure(figsize=(15, 10))
sns.countplot(x="Year", data=df, order = df.groupby(by=['Year'])['Video_Game'].count().sort_values(ascending=False).index)
plt.xticks(rotation=90);
```

```python
# games sales values worldwide

data_year = df.groupby(by=['Year'])['World Sales'].sum()
data_year = data_year.reset_index()

plt.figure(figsize=(15, 10))
sns.barplot(x="Year", y="World Sales", data=data_year)
plt.xticks(rotation=90);
```

```python
# sale price of each games' genre

data_genre = df.groupby(by=['Genre'])['World Sales'].sum()
data_genre = data_genre.reset_index()
data_genre = data_genre.sort_values(by=['World Sales'], ascending=False)

plt.figure(figsize=(15, 10))
sns.barplot(x="Genre", y="World Sales", data=data_genre)
plt.xticks(rotation=90);
```

```python
# sale price of each platfrom

data_platform = df.groupby(by=['Platform'])['World Sales'].sum()
data_platform = data_platform.reset_index()
data_platform = data_platform.sort_values(by=['World Sales'], ascending=False)

plt.figure(figsize=(15, 10))
sns.barplot(x="Platform", y="World Sales", data=data_platform)
plt.xticks(rotation=90);
```

```python
# sales compearison of each platforms

comp_platform = df[['Platform', 'North American Sales', 'EU Sales', 'Japan Sales', 'Other Sales']]
comp_platform.head()

comp_platform = comp_platform.groupby(by=['Platform']).sum().reset_index()

comp_table = pd.melt(comp_platform, id_vars=['Platform'], value_vars=[ 'North American Sales', 'EU Sales', 'Japan Sales', 'Other Sales'], var_name='Sale_Area', value_name='Sale_Price')
comp_table.head()

plt.figure(figsize=(30, 15))
sns.barplot(x='Platform', y='Sale_Price', hue='Sale_Area', data=comp_table)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
# list of publisher worldwide

sale_pbl = df[['Publisher', 'World Sales']]
sale_pbl = sale_pbl.groupby('Publisher')['World Sales'].sum().sort_values(ascending=False).head(20)
sale_pbl = pd.DataFrame(sale_pbl).reset_index()

plt.figure(figsize=(15, 10))
sns.barplot(x='Publisher', y='World Sales', data=sale_pbl)
plt.xticks(rotation=90);
```

```python
# total revenue

top_sale_reg = df[['North American Sales', 'EU Sales', 'Japan Sales', 'Other Sales']]
top_sale_reg = top_sale_reg.sum().reset_index()
top_sale_reg = top_sale_reg.rename(columns={"index": "region", 0: "sale"})
top_sale_reg

plt.figure(figsize=(12, 8))
sns.barplot(x='region', y='sale', data = top_sale_reg);
```
