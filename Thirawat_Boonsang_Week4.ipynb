```python
# EX 2
def sol(n):
  num = 0
  for i in range(1,n+1):
    if(n%i == 0):
      print(i,"*",n//i)
      num+=1 

  print("Total order =", num)

sol(8)
```

```python
# EX 3
arr=[]

for i in range (1, 4):
    for j in range (1, 3):
        element = 1/2*abs(i-(3*j))
        arr.append(element)

reMtx = np.reshape(arr,(3,2))
matrix32 = np.matrix(reMtx)
print(matrix32)
```

```python
# EX 4
A = np.matrix([[1, 2, 3], [2, 3, 1]])
B = np.matrix([[3, -1, 3], [-1, 0, 2]])

ans = (2*A)-B
print(ans)
```

```python
# EX 5
a = np.matrix([[5, 2], [0, 9]]) 
b = np.matrix([[3, 6], [0, -1]])

Y = (a-b)*0.5
X = np.subtract(a, Y)

print("Matrix X")
print(X, "\n")
print("Matrix Y")
print(Y, "\n")
```

```python
# EX 6
Factory_A = np.matrix([[80, 60],
                [75, 65],
                [90, 85]])

Factory_B = np.matrix([[90,50],
                [70,55],
                [75,75]])

t = Factory_A + Factory_B

c1 = np.sum(t[0])
c2 = np.sum(t[1])
c3 = np.sum(t[2])

print("Total : ")
print(t, "\n")
print("category1 :", c1)
print("category2 :", c2)
print("category3 :", c3)
```
