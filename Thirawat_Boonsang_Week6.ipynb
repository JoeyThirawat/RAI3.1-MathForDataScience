```python
# 4
x,y,c,m = sp.symbols('x,y,c,m')
exp = ((3*x)-1)*sp.log(2-x)
diff = sp.diff(exp, x)
print('derivative w.r.t x: ', diff)

# at x=1
point_unknown = sp.solve([sp.Eq(diff,m), sp.Eq(x,1)])
print("find m: ", point_unknown)

# linear
# y = mx+c
# lin = (point_unknown[0][m]*x)+c
# find c
xy_coor = sp.solve([sp.Eq(exp,y), sp.Eq(x,1)])
print("xy coordinate: ", xy_coor)

# 0 = (-2*1) + c
find_c = sp.solve([sp.Eq(y,(point_unknown[0][m]*xy_coor[0][x])+c), sp.Eq(y,0)])
print("c val: ",find_c[c])

tangent = point_unknown[0][m]*x + find_c[c]
print("tangent: ", tangent)
```

```python
p1 = plot(exp, show = False)
p2 = plot(tangent, show = False, line_color = 'r')
p1.append(p2[0])

# p1
p1.show()
print("tangent line => red line")
```

```python
x,y,c = sp.symbols('x,y,c')
fn = x*(sp.sqrt(x-1))
df = sp.Derivative(fn, x)
ans = df.doit()

#at x=5

#find slope
tan_m = ans.subs(x,5)

#normal is the line that cross to tangent 90 degrees
#normal slop = -1/m
nor_m = -1/tan_m

#y=mx+c
#y=(-4/13)x+c
#x = 5
#y = ?
sol_y = fn.subs(x,5)


sol_c = sp.solve([sp.Eq(y,(nor_m*5)+c), sp.Eq(y,10)])

normal = nor_m*x+sol_c[c]

print("Gradient function dy/dx = ", ans)
print("x coordinate = 5")
print("tangent slope m = ", tan_m)
print("normal slop m = ", nor_m)
print("y coordinate = ", sol_y)
print("10 = (-4/13)(5) + c")
print("at (5,0) and m = -4/13, we get c = ", sol_c)
print("Normal = ", normal)
```
