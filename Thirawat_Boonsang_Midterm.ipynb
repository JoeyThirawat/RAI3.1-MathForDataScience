```python
# Libraries
import numpy as np
import sympy as sp
from sympy import *
from sympy.plotting import plot3d_parametric_line
```

```python
# Q.1 code
'''
note
- parametric eq.
- find the intersection point at x = 0, y = 0, z = 0
'''
A = [1,1,1]
B = [2,3,2]

q1_vector = [A[0]-B[0] ,A[1]-B[1], A[2]-B[2]]

print(q1_vector)
```

```python
x, y, z = sp.symbols('x, y, z')
t = sp.symbols('t')

# Symbol ==> python
# sp.symbols ==> from sympy

x = A[0] + t*q1_vector[0]
y = A[1] + t*q1_vector[1]
z = A[2] + t*q1_vector[2]
print(x)
print(y)
print(z)
```

```python
graph = [x,y,z]
plot_graph_q1 = plot3d_parametric_line(*graph)
```
